```python
import os
import requests
import pandas as pd
from bs4 import BeautifulSoup

companies = ['Telsa', 'TCS', 'Bitcoin', 'Etherium', 'Nyka', 'LIC', 'Axis Blue Chip Mutual Fund']

for company in companies:

    topic = company.replace(' ','+')
    link  = 'https://www.google.com/search?q=' + topic + '&&tbm=nws'
    res   = requests.get(link)
    soup  = BeautifulSoup(res.text, 'html.parser')
    
    news = []
    link = []
    for i in soup.find_all('a')[17:-6]:
        news.append(i.text)
        link.append(i.get('href')[7:-88])
    link = list(set(link))
    news_fin = []
    for i in range(len(news)):
        if(i%2 == 0):
            news_fin.append(news[i])

    print(company)
    for i in range(len(link)):
        print(news_fin[i])
#        os.system('say '+ news_fin[i])
#        print(link[i])
    print("-"*100)
```

```python
news_fin = pd.Series(news_fin)
link     = pd.Series(link )

df = pd.concat([news_fin,link], axis = 1)
```

```python
df.to_csv('news.csv')
```
